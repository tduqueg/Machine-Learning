# Regresión lineal simple

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('Salary_Data.csv')
X = dataset.iloc[:, :-1].values
Y = dataset.iloc[:, -1].values
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0)
```

Entrenar el modelo de regresión lineal simple con el training set

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
# Entrenamos el modelo con la función fit que recibe como parámetros los datos de entrenamiento (X_train, Y_train) 
regressor.fit(X_train, Y_train)

```

Predicción de resultados

```python
#La función predict recibe como parámetro los datos de test y devuelve las predicciones de los datos de test 
#y se almacenan en la variable Y_pred
Y_pred = regressor.predict(X_test)
```

Visualización de los resultados del set de entrenamiento


```python
#La función scatter recibe como parámetros los datos de entrenamiento (X_train, Y_train) y el color de los puntos
#que se van a mostrar en el gráfico (color = 'red')
plt.scatter(X_train, Y_train, color = 'red')
#La función plot recibe como parámetros los datos de entrenamiento (X_train, Y_train) y el color de la línea
#que se va a mostrar en el gráfico (color = 'blue')
#Scatter y plot se diferencian en que scatter muestra los puntos y plot muestra una línea
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.title('Salario vs Años de experiencia (Set de entrenamiento)')
plt.xlabel('Años de experiencia')
plt.ylabel('Salario')
plt.show()
```

Visualización de los resultados del test set

```python
plt.scatter(X_test, Y_test, color = 'red')
#No se cambian los datos de entrenamiento por los de test porque la línea de regresión es la misma para ambos
plt.plot(X_train, regressor.predict(X_train), color = 'blue')
plt.title('Salario vs Años de experiencia (Set de test)')
plt.xlabel('Años de experiencia')
plt.ylabel('Salario')
plt.show()
```
